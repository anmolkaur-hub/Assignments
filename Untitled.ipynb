```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python

# Loading the raw dataset

 
df = pd.read_csv('raw_customer_data.csv')
```

```python
df.head(5)
```

```python
print("\nShape of dataset:", df.shape)
print("\nColumn names:", list(df.columns))
print("\nData types:\n", df.dtypes)
print("\nMissing values in each column:\n", df.isnull().sum())
print("\nTotal duplicates:", df.duplicated().sum())
print("\nBasic stats:\n", df.describe())
```

```python

#  Handling  Duplicates

 

print(" Removing Duplicates")

 
print("Rows before removing duplicates:", len(df))
df = df.drop_duplicates()
print("Rows after removing duplicates:", len(df))
```

```python
# Fix Outliers in Age
 

print(" Fixing Outliers in Age column")

print("Age values before fixing:\n", df['Age'].value_counts().sort_index())
 
# Any age below 18 or above 100 is not valid
bad_ages = df[~df['Age'].between(18, 100)]
print("\nRows with bad age values:")
print(bad_ages[['Name', 'Age']])
 
df['Age'] = df['Age'].apply(lambda x: np.nan if x < 18 or x > 100 else x)
print("\nAge values after fixing outliers:")
print("Min Age:", df['Age'].min(), "  Max Age:", df['Age'].max())
```

```python

# Handle Missing Values

 

print("Filling Missing Values")

 
print("Missing values BEFORE filling:\n", df.isnull().sum())
 
# Fill Age with median (better than mean for skewed data)
age_median = df['Age'].median()
df['Age'] = df['Age'].fillna(age_median)
print(f"\nFilled missing Age with median: {age_median}")
 
# Fill Satisfaction with median
sat_median = df['Satisfaction'].median()
df['Satisfaction'] = df['Satisfaction'].fillna(sat_median)
print(f"Filled missing Satisfaction with median: {sat_median}")
 
# Fill Gender and City with mode (most common value)
gender_mode = df['Gender'].mode()[0]
city_mode   = df['City'].mode()[0]
df['Gender'] = df['Gender'].fillna(gender_mode)
df['City']   = df['City'].fillna(city_mode)
print(f"Filled missing Gender with mode: {gender_mode}")
print(f"Filled missing City with mode: {city_mode}")
 
print("\nMissing values AFTER filling:\n", df.isnull().sum())
 
```

```python

#  Standardize Text Columns

 
print("\n" + "=" * 50)
print("Standardizing Text (Gender, City)")
print("=" * 50)
 
print("Unique Gender values before:", df['Gender'].unique())
 
df['Gender'] = df['Gender'].str.strip().str.lower()
df['Gender'] = df['Gender'].replace({'m': 'male', 'f': 'female'})
df['Gender'] = df['Gender'].str.capitalize()
 
df['City'] = df['City'].str.strip().str.capitalize()
 
print("Unique Gender values after:", df['Gender'].unique())
print("Unique City values after:",   df['City'].unique())
 
 
```

```python

#  Clean data summary

 

print(" Cleaned Data Summary")

print(df.head(10).to_string())
print("\nFinal shape:", df.shape)
print("\nFinal stats:\n", df.describe())
 
```

```python

#  Key Insights before plotting

 

print("Quick Insights")

print("Average Age:            ", round(df['Age'].mean(), 1))
print("Average Monthly Income: ", round(df['Monthly_Income'].mean(), 1))
print("Average Amount Spent:   ", round(df['Spent_Last_Month'].mean(), 1))
print("Average Satisfaction:   ", round(df['Satisfaction'].mean(), 2))
print("\nSpending by Gender:")
print(df.groupby('Gender')['Spent_Last_Month'].mean().round(0))
print("\nSpending by City:")
print(df.groupby('City')['Spent_Last_Month'].mean().round(0).sort_values(ascending=False))
```

```python
 

# Visualizations

 
fig, axes = plt.subplots(2, 3, figsize=(16, 10))
fig.suptitle('Customer Data Analysis - Cleaned Dataset', fontsize=16, fontweight='bold', y=1.01)
 
# --- Plot 1: Age Distribution ---
axes[0, 0].hist(df['Age'], bins=15, color='steelblue', edgecolor='white')
axes[0, 0].set_title('Age Distribution')
axes[0, 0].set_xlabel('Age')
axes[0, 0].set_ylabel('Number of Customers')
axes[0, 0].axvline(df['Age'].mean(), color='red', linestyle='--', label=f"Mean = {df['Age'].mean():.1f}")
axes[0, 0].legend()




# --- Plot 2: Gender Count ---
gender_counts = df['Gender'].value_counts()
axes[0, 1].bar(gender_counts.index, gender_counts.values,
               color=['steelblue', 'salmon', 'gray'], edgecolor='white')
axes[0, 1].set_title('Customers by Gender')
axes[0, 1].set_xlabel('Gender')
axes[0, 1].set_ylabel('Count')
for i, v in enumerate(gender_counts.values):
    axes[0, 1].text(i, v + 0.3, str(v), ha='center', fontweight='bold')
 
# --- Plot 3: Satisfaction Score Bar Chart ---
sat_counts = df['Satisfaction'].value_counts().sort_index()
axes[0, 2].bar(sat_counts.index.astype(int), sat_counts.values,
               color='mediumpurple', edgecolor='white')
axes[0, 2].set_title('Satisfaction Score Distribution')
axes[0, 2].set_xlabel('Score (1 = Very Low, 5 = Very High)')
axes[0, 2].set_ylabel('Number of Customers')
axes[0, 2].set_xticks([1, 2, 3, 4, 5])
 
# --- Plot 4: Monthly Income vs Amount Spent ---
colors = {'Male': 'steelblue', 'Female': 'salmon'}
for gender, group in df.groupby('Gender'):
    c = colors.get(gender, 'gray')
    axes[1, 0].scatter(group['Monthly_Income'], group['Spent_Last_Month'],
                       label=gender, alpha=0.7, color=c)
axes[1, 0].set_title('Monthly Income vs Amount Spent')
axes[1, 0].set_xlabel('Monthly Income (₹)')
axes[1, 0].set_ylabel('Amount Spent (₹)')
axes[1, 0].legend()
 
# --- Plot 5: Average Spending by City ---
city_spend = df.groupby('City')['Spent_Last_Month'].mean().sort_values(ascending=True)
axes[1, 1].barh(city_spend.index, city_spend.values, color='teal', edgecolor='white')
axes[1, 1].set_title('Average Spending by City')
axes[1, 1].set_xlabel('Avg Amount Spent (₹)')
for i, v in enumerate(city_spend.values):
    axes[1, 1].text(v + 50, i, f'₹{v:,.0f}', va='center', fontsize=8)
 
# --- Plot 6: Correlation Heatmap ---
numeric_cols = df[['Age', 'Monthly_Income', 'Spent_Last_Month', 'Satisfaction']]
corr_matrix  = numeric_cols.corr()
sns.heatmap(corr_matrix, annot=True, fmt='.2f', cmap='coolwarm',
            ax=axes[1, 2], linewidths=0.5)
axes[1, 2].set_title('Correlation Heatmap')
 
plt.tight_layout()
plt.show()

plt.close()
```
